**Artificial/Deep Neural Network**



```python
import numpy as np 
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
dataset.head()
```

```python
# split the data into dependent and independent variable
x = dataset.iloc[:,3:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
x.shape
```

```python
y.head()
```

```python
x = pd.get_dummies(x, columns=['Geography', 'Gender'])
x
```

```python
x = x.drop(['Geography_France','Gender_Female'], axis = 1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_stand = sc.fit_transform(x)
x_stand
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_stand, y, test_size=0.2, random_state=1)
```

# Building Deep Neural Network (MultiLayerPerceptron)

```python
from tensorflow import keras
```

```python
# creat APT
dnn = tf.keras.models.Sequential()
#Adding first Hidden Layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))           # in 1st hidden layer i am giving 6 neuron

#Adding second Hidden Layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))           # units should be equal or less than hidden layer 1

# Tensorflow is a backend language, keras is a frontend language, layers means hidden layer
# Dense means fully connected ie., each input data is individually connected to each layers/neuron and we are using relu as a activation functions

# adding output
dnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))         # if it is binary class classification we can use sigmoid, if it is multiclass classification can use softMax because it gives multipule probability values

# Train the model
dnn.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])   # since it is a binary classification we use binary crossentropy and we evaluate using accuracy mostly

# apply training dataset and fit the model
dnn.fit(x_train, y_train, batch_size = 32, epochs=100)
```

```python
dnn.fit(x_train, y_train, batch_size = 32, epochs=100)
```

```python
# Predict the test result
y_pred = dnn.predict(x_test)
```

```python
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
